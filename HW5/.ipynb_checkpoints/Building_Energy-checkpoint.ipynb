```python
#many columns have non-float inputs; need to create function to convert 
#data to more usable format (floats)
#Fs function converts data to boolean array - true if float, false is not. pass boolean array through original array (as ~index) to return only float values
#use try and except - allows code to continue without crashing if try condition not met
#replace all non-relevant cell inputs as NaN values
def canconvert(mydata):
    try :
        float(mydata)
        return True
    except ValueError:
        return False

convert = np.vectorize(canconvert)
```

```python
#adding to convert data in necessary columns to floats

index_BBL = convert(nrg['BBL'])
nrg['BBL'][~index_BBL] = float('NaN')
nrg['BBL'] = [float(x) for x in nrg['BBL']]

index_EUI = convert(nrg['EUI'])
nrg['EUI'][~index_EUI] = float('NaN')
nrg['EUI'] = [float(x) for x in nrg['EUI']]

index_TFA = convert(nrg['Total_floor_area'])
nrg['Total_floor_area'][~index_TFA] = float('NaN')
nrg['Total_floor_area'] = [float(x) for x in nrg['Total_floor_area']]

index_PBBL = convert(bsize['BBL'])
bsize['BBL'][~index_PBBL] = float('NaN')
bsize['BBL'] = [float(x) for x in bsize['BBL']]

index_units = convert(bsize['UnitsTotal'])
bsize['UnitsTotal'][~index_units] = float('NaN')
bsize['UnitsTotal'] = [float(x) for x in bsize['UnitsTotal']]

#print bsize['UnitsTotal']
```
